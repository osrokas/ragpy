# Chroma DB exploration

This notebooks provides examples how to query explore embeddings stored in Chroma DB.

## Setup
 - Import libraries
 - Initialize Chroma DB client and collection

```python
import chromadb
import pandas as pd
```

```python
# Create a Chroma DB client
client = chromadb.PersistentClient(path="../chroma_db")
```

Now we get existing collections of Chroma database.

```python
# List collections
collections = client.list_collections()
collections
```

```python
# Load collection
collection = client.get_collection("langchain")
collection
```

```python
"Count:", collection.count()
```

Retrieve example data from collection

```python
data = collection.peek(limit=5)
```
